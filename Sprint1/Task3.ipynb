```
public class Product
{
    private string _name;
    private double _price;
    private string _manufacturer;

    // Конструктор
    public Product(string name, double price, string manufacturer)
    {
        Name = name;
        Price = price;
        Manufacturer = manufacturer;
    }

    public string Name
    {
        get { return _name; }
        set { _name = value; }
    }

    public double Price
    {
        get { return _price; }
        set { _price = value; }
    }

    public string Manufacturer
    {
        get { return _manufacturer; }
        set { _manufacturer = value; }
    }

    public virtual string GetInfo()
    {
        return $"Название: {Name}; Цена: {Price}; Производитель: {Manufacturer}";
    }

    public virtual void Discount(double discountPercentage)
    {
        if (discountPercentage >= 0 && discountPercentage <= 100)
        {
            Price -= Price * (discountPercentage / 100);
        }
    }

    public virtual void Display()
    {
        Console.WriteLine(GetInfo());
    }
}

public class Electronics : Product
{
    private int _warrantyPeriod;

    // Конструктор
    public Electronics(string name, double price, string manufacturer, int warrantyPeriod) 
        : base(name, price, manufacturer)
    {
        WarrantyPeriod = warrantyPeriod;
    }

    public int WarrantyPeriod
    {
        get { return _warrantyPeriod; }
        set { _warrantyPeriod = value; }
    }

    public override void Discount(double discountPercentage)
    {
        if (2 <= WarrantyPeriod && WarrantyPeriod <= 5)
        {
            discountPercentage -= discountPercentage / 2;
        }

        if (WarrantyPeriod > 5)
        {
            discountPercentage -= discountPercentage / 1.5;
        }

        if (discountPercentage >= 0 && discountPercentage <= 100)
        {
            Price -= Price * (discountPercentage / 100);
        }
    }
}

public class Clothing : Product
{
    private double _size;

    // Конструктор
    public Clothing(string name, double price, string manufacturer, double size) 
        : base(name, price, manufacturer)
    {
        Size = size;
    }

    public double Size
    {
        get { return _size; }
        set { _size = value; }
    }

    public override void Display()
    {
        Console.WriteLine($"{GetInfo()}; Размер: {Size}");
    }
}

public class Books : Product
{
    private string _author;

    // Конструктор
    public Books(string name, double price, string manufacturer, string author) 
        : base(name, price, manufacturer)
    {
        Author = author;
    }

    public string Author
    {
        get { return _author; }
        set { _author = value; }
    }

    public override string GetInfo()
    {
        return $"Название: {Name}; Цена: {Price}; Производитель: {Manufacturer}; Автор: {Author}";
    }
}

// Новый класс для взаимодействия объектов - Корзина покупок
public class ShoppingCart
{
    private List<Product> _products;

    public ShoppingCart()
    {
        _products = new List<Product>();
    }

    public void AddProduct(Product product)
    {
        _products.Add(product);
        Console.WriteLine($"Товар '{product.Name}' добавлен в корзину");
    }

    public void RemoveProduct(Product product)
    {
        if (_products.Remove(product))
        {
            Console.WriteLine($"Товар '{product.Name}' удален из корзины");
        }
    }

    public void DisplayCart()
    {
        Console.WriteLine("\n Корзина");
        if (_products.Count == 0)
        {
            Console.WriteLine("Корзина пуста");
            return;
        }

        double totalPrice = 0;
        for (int i = 0; i < _products.Count; i++)
        {
            Console.Write($"{i + 1}. ");
            _products[i].Display();
            totalPrice += _products[i].Price;
        }

        Console.WriteLine($"\nОбщая стоимость: {totalPrice:F2} руб.");
        Console.WriteLine($"Количество товаров: {_products.Count + 1}");
    }

    public void ApplyDiscountToAll(double discountPercentage)
    {
        Console.WriteLine($"\nПрименяем скидку {discountPercentage}% ко всем товарам в корзине:");
        foreach (var product in _products)
        {
            double oldPrice = product.Price;
            product.Discount(discountPercentage);
            Console.WriteLine($"{product.Name}: {oldPrice:F2} -> {product.Price:F2} руб.");
        }
    }
}




Console.WriteLine("Магазин электроники:");
Electronics laptop = new Electronics("MagicBook X16 Plus", 64490, "Honor", 1);
laptop.Discount(19);
laptop.Display();

Console.WriteLine("\nМагазин одежды:");
Clothing hoody = new Clothing("худи Casual", 4599, "Befree", 50);
hoody.Discount(75);
hoody.Display();

Console.WriteLine("\nКнижный магазин:");
Books detective = new Books("Мистер Мерседес", 254, "Ast", "Стивен Кинг");
detective.Discount(0);
detective.Display();

// Демонстрация взаимодействия объектов через корзину
Console.WriteLine("\n" + new string('=', 50));
Console.WriteLine("Взаимодействие объектов:");
Console.WriteLine(new string('=', 50));

ShoppingCart cart = new ShoppingCart();

// Создаем дополнительные товары
Electronics smartphone = new Electronics("iPhone 15", 89990, "Apple", 2);
Clothing tshirt = new Clothing("Футболка Basic", 1999, "Uniqlo", 48);
Books novel = new Books("1984", 450, "Penguin", "Джордж Оруэлл");

// Добавляем товары в корзину
cart.AddProduct(laptop);
cart.AddProduct(hoody);
cart.AddProduct(smartphone);
cart.AddProduct(tshirt);
cart.AddProduct(novel);

// Показываем корзину
cart.DisplayCart();

// Применяем скидку
cart.ApplyDiscountToAll(10);

// Показываем корзину после скидки
cart.DisplayCart();

// Удаляем товар и показываем корзину
Console.WriteLine("\nУдаляем товар из корзины:");
cart.RemoveProduct(hoody);
cart.DisplayCart();

Console.WriteLine("\n");

   



```
